```python
from qiskit import QuantumCircuit, QuantumRegister
q = QuantumRegister(5)
qc = QuantumCircuit(q)
```

```python
qc.h(q[0])
qc.h(q[4])
qc.mct([q[0],q[1],q[3]],q[4])
qc.swap(q[2],q[3])
qc.cy(q[0],q[4])
qc.draw('mpl')
```

```python
inverse_circuit = qc.inverse()
```

```python
inverse_circuit.draw('mpl')
```

```python
my_gate = qc.to_gate()
```

```python
q = QuantumRegister(7)
qc1 = QuantumCircuit(q)
```

```python
qc1.append(my_gate,[q[1],q[2],q[3],q[4],q[5]])
```

```python
qc1.draw('mpl')
```
