```python
from qiskit import QuantumRegister, QuantumCircuit
q = QuantumRegister(5)
qc = QuantumCircuit(q)
```

```python
from qiskit.quantum_info import Statevector
```

```python
psi1 = Statevector(qc)
psi1.draw('latex')
```

```python
qc.x(q[0])
qc.x(q[1])
qc.ccx(q[0],q[1],q[2])
qc.draw('mpl')
```

```python
psi2 = Statevector(qc)
psi2.draw('latex')
```

```python
qc.mct([q[0],q[2],q[3]],q[4])
qc.draw('mpl')
```

```python
from math import pi
```

```python
qc.mcry(pi/2,[q[1],q[2],q[3],q[4]],q[0])
qc.draw('mpl')
```
